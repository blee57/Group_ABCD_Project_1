```python
import pandas as pd
```

# cleaning data for dropout rate and general academic performance for merge with performance data

```python
# importing the dataset
df_dropout = pd.read_csv('../dataset/7-12 Dropout Rate-2018_DEC_10th_2018.csv')
df_general = pd.read_csv('../dataset/district-18.csv')
```

```python
df_general.head()
```

```python
df_gen_col = pd.DataFrame
df_gen_col = df_general.loc[:,'pct_asian':'Four_Year_Graduation_Rate']
df_gen_col['SystemID'] = df_general.loc[:,'SystemId']
```

```python
df_gen_col = df_gen_col.rename(columns={'SystemID' : 'School District Code'})
df_gen_col['School District Code']
df_gen_col ['School District Code'] = df_gen_col['School District Code'].astype(int)
df_gen_col = df_gen_col.dropna().reset_index(drop = True)
```

```python
df_gen_col.to_csv('../data_export/general_performance.csv')
```

```python
df_dropout
```

```python
df_dropout = df_dropout[[ 'SCHOOL_DSTRCT_CD',
       'LABEL_LVL_1_DESC', 'PROGRAM_TOTAL', 'PROGRAM_PERCENT']]
```

```python
df_drop_id = df_dropout.groupby('SCHOOL_DSTRCT_CD').mean().dropna().iloc[:-1]
```

```python
df_drop_id = df_drop_id.reset_index()
```

```python
df_drop_id = df_drop_id.rename(columns={'SCHOOL_DSTRCT_CD':'School District Code'})
df_drop_id['School District Code'] = df_drop_id['School District Code'].astype(int)
```

```python
drop_clean_data = df_drop_id.copy()
```

```python
drop_clean_data.to_csv('../data_export/drop_out_clean.csv')
```
