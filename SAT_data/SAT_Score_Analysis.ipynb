```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
```

```python
# Read in SAT Score CSV
sat_scores_df=pd.read_csv('SAT-score-data.csv')
sat_scores_df
```

```python
# Read graduation rate CSV
graduation_rate_df = pd.read_csv("../data_export/school_clean.csv")
graduation_rate_df
```

```python
# Merge dataframes
grad_sat_df = pd.merge(sat_scores_df, graduation_rate_df, on="School District Code")
grad_sat_df
```

```python
clean_grad_sat_df=grad_sat_df.drop(columns=['School Year', 'School District Name','School Code','School Name'])
clean_grad_sat_df
```

```python
# Compare average SAT score with graduation rate (scatterplot/linear regression)
clean_grad_sat_df.plot(kind="scatter", x="District Average Score", y="Graduates", grid=True, figsize=(8,8),
              title="Average SAT Score vs Number of Graduates")
plt.show()
```

```python
# Linear Regression
x_values = clean_grad_sat_df['District Average Score']
y_values = clean_grad_sat_df['Graduates']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel('Graduates')
print(f"The r-squared is: {rvalue**2}")
plt.show()
```

```python
# Compare average SAT score with college enrollment (scatterplot/linear regression)
clean_grad_sat_df.plot(kind="scatter", x="District Average Score", y="Postsecondary Institution", grid=True, figsize=(8,8),
              title="Average SAT Score vs College Enrollment")
plt.show()
```
