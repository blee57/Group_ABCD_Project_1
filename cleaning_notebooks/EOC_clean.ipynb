```python
import pandas as pd
```

```python
df_eoc = pd.read_csv('../dataset/EOC_2019_By_Grad_FEB_24_2020 (1).csv')
```

```python
df_id_subgroup = df_eoc.groupby(['SCHOOL_DISTRCT_CD', 'SUBGROUP_NAME']).mean().reset_index()
df_id_sub_noall = df_id_subgroup.loc[df_id_subgroup['SUBGROUP_NAME'] != 'All Students']
```

```python
subgroup_clean = df_id_sub_noall.rename(columns=
    {
        'SCHOOL_DISTRCT_CD' : 'School District Code'
    }
)

```

```python
subgroup_clean.to_csv('../data_export/subgroup_EOC.csv')
```

```python
df_id_type = df_eoc.groupby(['SCHOOL_DISTRCT_CD', 'TEST_CMPNT_TYP_NM']).mean().reset_index()
```

```python
df_id_type.rename(columns={
    'SCHOOL_DISTRCT_CD' : 'School District Code'
})
```

```python
type_clean = df_id_type.copy()
type_clean
```
